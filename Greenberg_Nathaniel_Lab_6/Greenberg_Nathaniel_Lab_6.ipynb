```python
from selenium import webdriver #import everything that is neccesary 
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
import pandas as pd
import csv
import requests
import html5lib
import csv

PATH = Service("/Users/nategreenberg/Desktop/Harp 151/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)
driver.get("https://www.binghamton-ny.gov/home") #here i go to the website

gov = driver.find_element(By.XPATH,'//*[@id="dropdownrootitem3"]/a')
dep = driver.find_element(By.XPATH,'//*[@id="dropdownrootitem3"]/div/div/ul[1]/li/a')

actions = ActionChains(driver) #here i use an action chain to navigate 
actions.move_to_element(gov)
actions.click(gov)
actions.move_to_element(dep)
actions.click(dep)
actions.perform()

csv_file = open("job_scrape.csv", "w", newline="", encoding="utf-8") #here I create the csv
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Job","Type","Application Deadline","Salary"])


try:
    click_xpath = '//*[@id="widget_4_33_127"]/ul/li[16]/a' #here i use try/finally and webdriverwait to navigate
    link = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, click_xpath)))
    link.click()
    click_xpath = '//*[@id="leftNav_1038_0_145"]/ul/li/ul/li[14]/ul/li/a'
    link = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, click_xpath)))
    link.click()
    
    main = WebDriverWait(driver,10).until(
    EC.presence_of_element_located((By.XPATH,'//*[@id="ColumnUserControl4"]/div[2]/table')))
    table = main.find_elements(By.TAG_NAME,"tr")
    num = 0
    
    for i in table:
        job = i.find_element(By.TAG_NAME, "a").text
        deadline = i.find_element(By.CLASS_NAME,"mobile_hide").text
        if num > 0: #using number variable and f string to change the xcode of the table since it wasn't iterating well
            salary = i.find_element(By.XPATH, f'//*[@id="ColumnUserControl4"]/div[2]/table/tbody/tr[{num}]/td[4]').text
            job_type = i.find_element(By.XPATH, f'//*[@id="ColumnUserControl4"]/div[2]/table/tbody/tr[{num}]/td[2]').text
        else: #This is for the headings
            salary = i.find_element(By.XPATH, '//*[@id="ColumnUserControl4"]/div[2]/table/thead/tr/th[4]/a').text
            job_type = i.find_element(By.XPATH, '//*[@id="ColumnUserControl4"]/div[2]/table/thead/tr/th[2]/a').text
        csv_writer.writerow([job,job_type,deadline,salary])
        num+= 1
        
finally:
        driver.quit()
        csv_file.close()
        
sheet = pd.read_csv("job_scrape.csv")
sheet  
```
