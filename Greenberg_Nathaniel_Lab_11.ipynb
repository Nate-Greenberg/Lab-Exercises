```python
import random
from tkinter import *
import requests
import json

#I got everything in the lab done correctly besides from one factor. The answer choice don't clear when I run a 
#new question. I tried using the destroy method but I couldn't get it to work. Besides the questions not clearing,
#the program still has a button to go to the next question, a counter and displays the number of correct/incorrect
#at the end. 

request = requests.get("https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy").json()

root = Tk()
root.title("Trivia Game")
root.geometry("1000x800")

score = 0
count = 0
total = 1

question_list = []
for i in request:
    i = i["question"]
    question_list.append(i)
random.shuffle(question_list)

canvas = Canvas(root, width=800, height=250, bg="lightblue")
canvas.place(relx=.5, rely=.5, anchor= CENTER) 

display = canvas.create_text(200, 75, text="", fill="black")

iscorrect = Label(root, text="", pady=10, font=("arial", 15, "bold"))

showscore = Label(root,text="",pady=10, font=("arial", 15, "bold"))

user_answer = StringVar()
user_answer.set("")

def question(q):
    global total
    if total < 10:
        showscore["text"] = f"Your score is {score}"
        showscore.place(x=410,y=530) 
    else:
        showscore["text"] = f"You got {score} correct and {10-score} incorrect"
        showscore.place(x=410,y=530)
    answer_list = []
    for i in request:
        if i["question"] == q:
            correct = i["correctAnswer"]
            answer_list.append(correct)
            for incorrect in i["incorrectAnswers"]:
                answer_list.append(incorrect)
    random.shuffle(answer_list)

    canvas.itemconfig(display, text= q, width=300)
    


    def evaluate():
        global score
        answer = user_answer.get()
        for i in request:
            if i["question"] == q:
                if i["correctAnswer"] == answer:
                    iscorrect["text"] = "Your Right!"
                    iscorrect.place(x=400, y=500)
                    score += 1
                else:
                    iscorrect["text"] = "Your Wrong"
                    iscorrect.place(x=400, y=500)
                    
    y_pos = 100
    for i in answer_list:
        rad_btn = Radiobutton(canvas, text=i, variable=user_answer, command= evaluate, value=i, font=('Helvetica 10 bold'))
        rad_btn.place(x=50, y=y_pos)
        y_pos+=20
        
             
def next_button(): 
    global count
    global total
    count += 1
    total += 1
    button = Button(canvas,text="Press for next Question", command= lambda: [question(question_list[count]),next_button()])    
    button.place(x=50,y=0)
    answer_list = []
    
question(question_list[count])
next_button()
    
root.mainloop()
```
