```python
#Start of hoops hype web scrape 
from bs4 import BeautifulSoup
import pandas as pd
import html5lib
import requests
import csv

PATH = Service("/Users/nategreenberg/Desktop/Harp 151/chromedriver.exe") #here I set the path
driver = webdriver.Chrome(service=PATH) #Here I set the driver
hoop = "https://hoopshype.com/" #here i set the website link into a variable
driver.get(hoop)
link = driver.find_element(By.LINK_TEXT,"SALARIES") #locates the salary link
link.click() #here I click the link 
link = driver.find_element(By.LINK_TEXT,"PLAYER SALARIES") #locates the salary link
link.click() #here I click the link 

csv_file = open("nba_scrape.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Salary"])

try:
    main = WebDriverWait(driver, 10).until(  
    EC.presence_of_element_located((By.ID,"main")) 
    )
    table = main.find_elements(By.TAG_NAME,"tr")
    for i in table:
        Salary = i.find_element(By.CLASS_NAME,"hh-salaries-sorted") #locates entry-summary class 
        csv_writer.writerow([Salary.text])
finally:
        driver.quit()
        csv_file.close()
        
sheet=pd.read_csv("nba_scrape.csv")
sheet  
```
