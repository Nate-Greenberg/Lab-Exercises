```python
#Question one-
import requests 
import json
from tkinter import *

author_list = []
quote = requests.get("https://api.quotable.io/authors").json()
for i in quote["results"]:
    author_list.append(i["name"])

root = Tk()
root.title("Quote Dropdown")
root.geometry("1000x600")

def chosen_option():
    author = value.get()
    result_label = Label(root, text=f"You've chosen {author}")
    result_label.pack()
    quote_data = requests.get(f"https://api.quotable.io/quotes?author={author}").json()
    quote_list = []
    for i in quote_data["results"]:
        quote_list.append(i["content"])
    quote = (quote_list[0])
    quote_label = Label(root, text =f"Here is your quote: {quote}")
    quote_label.pack()
    
value = StringVar()
value.set("Select an Author")
    
dropdown = OptionMenu(root, value, *author_list)
choose_button = Button(root, text="Click to get quote", command=chosen_option)

dropdown.pack()
choose_button.pack()

root.mainloop()
```

```python
#Question two- 
latitude = 42.0987
longitude = -75.9180
json_file = requests.get(f"https://api.weather.gov/points/{latitude},{longitude}").json()

root = Tk()  
root.geometry("560x180") 
root.title("Forecast")

forecast = json_file["properties"]["forecast"]
newrequest = requests.get(forecast)
json = newrequest.json()
a = json["properties"]["periods"]
names = []
temps = []
details = []
for i in a:
    name=i["name"]
    temp=i["temperature"]
    detail=i["shortForecast"]
    names.append(name)
    temps.append(temp)
    details.append(detail)
x = 1
for i in range (7):
    frame = Frame(root)
    frame.pack()
    myLabel = Label(root, text=f"Day: {names[x]}, Temp = {temps[x]}ยบ, Details: {details[x]}")
    myLabel.pack()
    x += 2
    
root.mainloop()
```
